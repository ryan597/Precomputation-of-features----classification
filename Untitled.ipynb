```python
class dset(Dataset):
    def __init__(self, imgpath, labpath):
        self.imgpath = imgpath
        self.labpath = labpath

        self.img = None
        self.lab = None
        
        with h5py.File(self.labpath, 'r') as f:
            self.dataset_len = len(f['dataset_1'])

    def __getitem__(self, index):
        if self.img is None:
            self.img = h5py.File(self.imgpath, 'r')['dataset_1']
            self.lab = h5py.File(self.labpath, 'r')['dataset_1']
        
        return (self.img[index], self.lab[index])

    def __len__(self):
        return self.dataset_len
```

```python
class net(nn.Module):
    def __init__(self, in_features, mid_features, out_features):
        super().__init__()
        self.lin = nn.Sequential(
            nn.Linear(in_features, mid_features),
            nn.BatchNorm1d(mid_features),
            nn.LeakyReLU(),
            nn.Dropout(0.6),
            nn.Linear(mid_features, mid_features),
            nn.BatchNorm1d(mid_features),
            nn.LeakyReLU(),
            nn.Dropout(0.6),
            nn.Linear(mid_features, out_features)
        )
    def forward(self, x):
        return self.lin(x)
```

```python
n = net(4096, 40, 3)
#weights = find_weight(train)
weights = torch.FloatTensor([0, 1, 0.1])
print(weights)
criterion = nn.CrossEntropyLoss(weight=weights)
optimizer = optim.AdamW(n.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer,
                                                     T_max=100)
```

```python
scheduler.optimizer.zero_grad()
trainloss = []
valloss = []
for epoch in range(100):
    print(f"Epoch : {epoch}")
    n.train()
    for i, (img, lab) in enumerate(train):
        outputs = n(img)
        loss = criterion(outputs, lab)
        
        trainloss.append(loss.item())
        loss.backward()
        print(f"Loss : \t{loss.item()}")
        scheduler.optimizer.step()
        scheduler.optimizer.zero_grad()

    with torch.no_grad():
        n.eval()
        for i, (vimg, vlab) in enumerate(test):
            val = n(vimg)
            vloss = criterion(val, vlab)
            print(f"Validation loss :\t{vloss.item()}")
            valloss.append(vloss.item())
            #one_hot = np.zeros((len(val), 3))
            #for j in lab:
            #    one_hot[j] = 1

            print(classification_report(vlab.numpy(), np.argmax(val.numpy(),-1)))
    scheduler.step()
```

```python
plt.plot(trainloss, label = 'Train')
plt.plot(valloss, label='Val')
plt.legend()
```
